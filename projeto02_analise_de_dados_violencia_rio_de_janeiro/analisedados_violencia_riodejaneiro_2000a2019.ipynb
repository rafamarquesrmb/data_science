Vale lembrar que os dados obtidos nesse DataFrame seguem um período de longas datas. Durante esse processo diversas possibilidades podem justificar a ausência desses dados. Algumas possibilidades:
* Falha nos registros;
* Perda de informação ao longo dos anos;
* Perda de informação por modificação no método de armazenamento desses dados;
* Possivelmente alguns dados não eram coletados inicialmente, passando a ser contabilizados só a partir de determinado período;
* entre outros.

Dentre todas essas possibilidades, podemos analisar nossos dados para encontrar a mais provável e entender melhor o motivo dos dados faltantes.

### Número de Ocorrência por Ano e Mês

Nesta etapa, vamos verificar qual foi o ano que teve o maior número de ocorrências. Posteriormente vamos também verificar em qual mês costuma ter um maior numero de ocorrências. Para esse segundo caso, vamos fazer a média de registro para cada mês.

```
aux = df_xxi.loc[df_xxi['vano'] < 2019].groupby(['vano']).registro_ocorrencias.sum()
colori = []
for i in aux.values:
    if i >= (aux.values.max())*0.95 :
        colori.append('#34495e')
    else:
        colori.append('#95a5a6')
plt.figure(figsize=(13,5))
plt.bar(aux.index,aux.values,color=colori);
plt.xticks(aux.index , rotation=45)
plt.yticks([])
plt.ylabel('Total de Ocorrências')
plt.grid(False)
plt.title('Total de Ocorrências por Ano no Rio de Janeiro')
plt.tight_layout()
```

```
df_xxi.loc[df_xxi['vano'] < 2019].groupby(['vano']).registro_ocorrencias.sum().describe()
```

Vale ressaltar que o ano de 2019 apresenta registros apenas até o mês de agosto, portanto, esse não é o valor correspondente a todo o ano de 2019.

```
meses = pd.Series(['jan','fev','mar','abr','mai','jun','jul','ago','set','out','nov','dez'])
aux = df_xxi.groupby(['mes']).registro_ocorrencias.mean()
aux.index = meses
colori = []
for i in aux.values:
    if i >= (aux.values.max()):
        colori.append('#34495e')
    else:
        colori.append('#95a5a6')
plt.figure(figsize=(13,5))
plt.bar(aux.index,aux.values,color=colori);
plt.xticks(aux.index , rotation=45)
plt.yticks([])
plt.ylabel('Média de Ocorrências')
plt.grid(False)
plt.title('Médias de Ocorrências por Meses no Rio de Janeiro (2000 à 2019)')
plt.tight_layout()
```
